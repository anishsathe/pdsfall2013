```python
import pandas as pd
import pandas.tools.pivot as pt
import numpy as np
import matplotlib.pyplot as plt

survey = "/Users/anish_sathe/Documents/Data Mining/dsassignments/survey.csv"
df = pd.read_csv(survey,delimiter=',',header=None, \
                     names=['Timestamp','Degree Program','Unix Experience','Database Experience', 'Program Experience'])

dfunixexp = pd.DataFrame([3,4,1,2,5], index=['I have issued a few commands in a terminal based on given instructions ',
                                        'I have written simple terminal commands or done some system work on the terminal ',
                                        'I don\xe2\x80\x99t even understand the question',
                                        'I have no experience working in a terminal ',
                                        'I have written complex commands done or have done deep system work  '],
                       columns = ['Unix Lvl'])
dfdbexp = pd.DataFrame([3,4,1,2,5], index=['I can write simple queries and issue them to a database ',
                                         'I can write very complex queries when needed ',
                                         'I have never directly accessed a database ',
                                         'I have issued simple queries to a relational database based on given instructions ',
                                         'I am a database hacker '],
                       columns = ['DB Lvl'])
dfprgmexp = pd.DataFrame([2,3,1,4,5], index=['I have written simple programs based on instructions or a tutorial ',
                                         'I can write simple programs to accomplish tasks I encounter ',
                                         'I have never programmed before. ',
 'I can write complex programs am familiar with programming design patterns software testing system design and algorithms. ',
                                            'I am a hacker or have  senior-level programming experience '],
                       columns = ['Prgm Lvl'])

print '# of students responed:', df['Timestamp'].count()
print df.groupby('Unix Experience').size()
print df.groupby('Database Experience').size()
print df.groupby('Program Experience').size()

df = df.join(dfunixexp, on='Unix Experience', how='left', lsuffix='', rsuffix='_r', sort=False)
df = df.join(dfdbexp, on='Database Experience', how='left', lsuffix='', rsuffix='_r', sort=False)
df = df.join(dfprgmexp, on='Program Experience', how='left', lsuffix='', rsuffix='_r', sort=False)

dfu = pd.DataFrame(df.groupby(['Unix Lvl']).size(),columns=['Unix Lvl'])
dfb = pd.DataFrame(df.groupby(['DB Lvl']).size(),columns=['DB Lvl'])
dfp = pd.DataFrame(df.groupby(['Prgm Lvl']).size(),columns=['Prgm Lvl'])
p = dfu.join(dfb, how='outer', sort=False)
p = p.join(dfp, how='outer', sort=False)

p.plot(kind='line',linewidth=2.0)
p.plot(kind='bar')
plt.title("PDS Skill Survey")
plt.xlabel('Skill Lavel')
plt.ylabel('# Of Students')
plt.legend(loc='best')
plt.show()
```
